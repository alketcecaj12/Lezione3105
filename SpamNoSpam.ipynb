```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as pn
import sklearn
```

```python
training_set = pd.read_csv("dati_spam/spambase_trn.csv")
test_set = pd.read_csv("dati_spam/spambase_tst.csv")
```

```python
test_set
```

```python
X_train = training_set.iloc[:, :-1].values
y_train = training_set["y"].values
```

```python
X_test = test_set.iloc[:, :-1].values
y_test = test_set["y"].values
```

```python
from sklearn.neighbors import KNeighborsClassifier 
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train, y_train)
```

```python
previsione = knn.predict(X_test)
```

```python
from sklearn import metrics 
```

```python
metrics.accuracy_score(previsione, y_test)
```

#### Faccio scaling delle features 

```python
from sklearn import preprocessing
X_train = preprocessing.scale(X_train)
X_test = preprocessing.scale(X_test)
```

```python
knn2 = KNeighborsClassifier(n_neighbors=5)
```

```python
knn2.fit(X_train, y_train)
```

```python
previsione2 =  knn2.predict(X_test)
```

```python
metrics.accuracy_score(previsione2, y_test)
```

#### Cerco i parametri migliori 

```python
from sklearn.model_selection import GridSearchCV
```

```python
k_range = list(range(1,31))
grid_param = dict(n_neighbors = k_range)
grid = GridSearchCV(knn, grid_param, cv = 10, scoring= "accuracy", n_jobs=-1)
```

```python
grid.fit(X_train, y_train)
```

```python
results = grid.cv_results_
```

```python
results
```

```python
dati_ott = pd.DataFrame(results)
```

```python
dati2opt = dati_ott[["param_n_neighbors", "mean_test_score","std_test_score","rank_test_score"]]
```
